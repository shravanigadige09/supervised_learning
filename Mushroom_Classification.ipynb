# 2.Data loading and Description of Dataset
>The dataset for this project consists of various types of mushrooms and their properties. It includes features such as


*   **class:** This column likely indicates the classification of mushrooms as either edible or poisonous.

*   **cap-surface:** Indicates the surface texture of the mushroom cap (e.g., smooth, scaly).
*   **cap-color:** Represents the color of the mushroom cap.


*   **cap-shape:** Describes the shape of the mushroom cap (e.g., bell, conical, convex).

*   **bruises:** Likely specifies whether the mushroom bruises or not.

*   **odor:** Describes the odor of the mushroom (e.g., foul, almond, anise).

*   **gill-attachment:** Indicates how the gills are attached to the stem (e.g., free, attached).

*   **gill-spacing:** Represents the spacing between gills on the mushroom&#39;s underside (e.g., close,
crowded).

*   **gill-size:** Specifies the size of the gills (e.g., narrow, broad).
*   gill-color: Represents the color of the gills.


*   **stalk-shape:** Indicates the shape of the mushroom&#39;s stem (e.g., tapering, enlarging).


*   **stalk-root:** Describes the type of root base the mushroom has (e.g., bulbous, club, equal,
rooted).


*   **stalk-surface-above-ring:** Indicates the texture of the stem&#39;s surface above the ring (if present).


*   **stalk-surface-below-ring:** Represents the texture of the stem&#39;s surface below the ring (if
present).

*   **stalk-color-above-ring:** Describes the color of the stem above the ring (if present).
*   stalk-color-below-ring: Represents the color of the stem below the ring (if present).

*   **veil-type:** Indicates the type of veil covering the gills (this column may have the same value for
all records, as it is typically not a distinguishing feature).

*   **veil-color:** Specifies the color of the veil.
*   ring-number: Represents the number of rings on the stem (e.g., none, one, two).

*   **ring-type:** Describes the type of ring on the stem (e.g., pendant, evanescent, large, small).

*   **spore-print-color:** Represents the color of the spore print.
*   **population:** Indicates the population size of the mushroom (e.g., abundant, numerous,
scattered).

*   **habitat:** Describes the habitat where the mushroom is typically found (e.g., grasses, woods,
waste).

```python
# mapping the each column with related data for clear understanding

df.loc[:,'class'] = df['class'].map({'p':'poisonous','e':'edible'})
df.loc[:,'cap-shape'] = df['cap-shape'].map({'b':'bell','c':'conical','x':'convex','f':'flat','k':'knobbed','s':'sunken'})
df.loc[:,'cap-surface'] = df['cap-surface'].map({'f':'fibrous','g':'grooves','y':'scaly','s':'smooth'})
df.loc[:,'cap-color'] = df['cap-color'].map({'n':'brown','b':'buff','c':'cinnamon','g':'gray','r':'green','p':'pink','u':'purple','e':'red','w':'white','y':'yellow'})
df.loc[:,'bruises'] = df['bruises'].map({'t':'bruises','f':'no'})
df.loc[:,'odor'] = df['odor'].map({'a':'almond','l':'anise','c':'creosote','y':'fishy','f':'foul','m':'musty','n':'none','p':'pungent','s':'spicy'})
df.loc[:,'gill-attachment'] = df['gill-attachment'].map({'a':'attached','d':'descending','f':'free','n':'notched'})
df.loc[:,'gill-spacing'] = df['gill-spacing'].map({'c':'close','w':'crowded'})
df.loc[:,'gill-size'] = df['gill-size'].map({'b':'broad','n':'narrow'})
df.loc[:,'gill-color'] = df['gill-color'].map({'k':'black','n':'brown','b':'buff','h':'chocolate','g':'gray','r':'green','o':'orange','p':'pink','u':'purple','e':'red','w':'white','y':'yellow'})
df.loc[:,'stalk-shape'] = df['stalk-shape'].map({'e':'enlarging','t':'tapering'})
df.loc[:,'stalk-root'] = df['stalk-root'].map({'b':'bulbous','c':'club','u':'cup','e':'equal','z':'rhizomorphs','r':'rooted','?':'missing'})
df.loc[:,'stalk-surface-above-ring'] = df['stalk-surface-above-ring'].map({'f':'fibrous','y':'scaly','k':'silky','s':'smooth'})
df.loc[:,'stalk-surface-below-ring'] = df['stalk-surface-below-ring'].map({'f':'fibrous','y':'scaly','k':'silky','s':'smooth'})
df.loc[:,'stalk-color-above-ring'] = df['stalk-color-above-ring'].map({'n':'brown','b':'buff','c':'cinnamon','g':'gray','o':'orange','p':'pink','e':'red','w':'white','y':'yellow'})
df.loc[:,'stalk-color-below-ring'] = df['stalk-color-below-ring'].map({'n':'brown','b':'buff','c':'cinnamon','g':'gray','o':'orange','p':'pink','e':'red','w':'white','y':'yellow'})
df.loc[:,'veil-type'] = df['veil-type'].map({'p':'partial','u':'universal'})
df.loc[:,'veil-color'] = df['veil-color'].map({'n':'brown','o':'orange','w':'white','y':'yellow'})
df.loc[:,'ring-number'] = df['ring-number'].map({'n':'none','o':'one','t':'two'})
df.loc[:,'ring-type'] = df['ring-type'].map({'c':'cobwebby','e':'evanescent','f':'flaring','l':'large','n':'none','p':'pendant','s':'sheathing','z':'zone'})
df.loc[:,'spore-print-color'] = df['spore-print-color'].map({'k':'black','n':'brown','b':'buff','h':'chocolate','r':'green','o':'orange','u':'purple','w':'white','y':'yellow'})
df.loc[:,'population'] = df['population'].map({'a':'abundant','c':'clustered','n':'numerous','s':'scattered','v':'several','y':'solitary'})
df.loc[:,'habitat'] = df['habitat'].map({'g':'grasses','l':'leaves','m':'meadows','p':'paths','u':'urban','w':'waste','d':'woods'})
df
```
